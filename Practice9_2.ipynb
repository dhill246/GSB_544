
Our dataset consists of clinical data from patients who entered the hospital complaining of chest pain ("angina") during exercise.  The information collected includes:

* `age` : Age of the patient

* `sex` : Sex of the patient

* `cp` : Chest Pain type

    + Value 0: asymptomatic
    + Value 1: typical angina
    + Value 2: atypical angina
    + Value 3: non-anginal pain
   
    
* `trtbps` : resting blood pressure (in mm Hg)

* `chol` : cholesterol in mg/dl fetched via BMI sensor

* `restecg` : resting electrocardiographic results

    + Value 0: normal
    + Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    + Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

* `thalach` : maximum heart rate achieved during exercise

* `output` : the doctor's diagnosis of whether the patient is at risk for a heart attack
    + 0 = not at risk of heart attack
    + 1 = at risk of heart attack

```python
## library imports here
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.tree import plot_tree
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
ha = pd.read_csv("https://www.dropbox.com/s/aohbr6yb9ifmc8w/heart_attack.csv?dl=1")
```

```python
ha.head()
```

```python
# When I column transform chest pain and other categorical
# variables represented as numbers, I don't want to have
# to deal with interpretting the 1's, 2's and so on, I'd rather
# just interpret as is, so I'm transforming.

# Chest Pain Transformer
def output(x):
    if x == 0:
        return "NoRisk"
    else:
        return "Risk"

# Resting Electrocardiographic Results
def restecg(x):
    if x == 0:
        return "Normal"
    elif x == 1:
        return "STTWaveAbnormal"
    else:
        return "LeftVentHypertrophy"

# Sex
def sex(x):
    if x == 1:
        return "Male"
    else:
        return "Female"

# Chest Pain Transformer
def chest_pain(x):
    if x == 1:
        return "TypicalAngina"
    elif x == 2:
        return "AtypicalAngina"
    elif x == 3:
        return "NonAnginalPain"
    else:
        return "Asymptomatic"


ha["restecg"] = ha["restecg"].apply(restecg)
ha["sex"] = ha["sex"].apply(sex)
ha["output"] = ha["output"].apply(output)
ha["cp"] = ha["cp"].apply(chest_pain)
```

## Q1: Natural Multiclass Models

Fit a multiclass KNN, Decision Tree, and LDA for the heart disease data; this time predicting the type of chest pain (categories 0 - 3) that a patient experiences.  For the decision tree, plot the fitted tree, and interpret the first couple splits.


```python
X = ha.drop(columns=["cp"])
y = ha["cp"]
```

```python
def buildColTransformer():
    """
    Constructs a default column transformer for encoding categorical vars
    and scaling numeric vars

    Parameters
    ----------
    None

    Returns
    -------
    ct : sklearn ColumnTransformer
        ct object ready to be used in a sklearn Pipeline


    """
    enc = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
    scaler = StandardScaler()
    get_categoric = make_column_selector(dtype_include=object)
    get_numeric = make_column_selector(dtype_include=np.number)

    ct = ColumnTransformer([
        ("dummify", enc, get_categoric),
        ("standardize", scaler, get_numeric)
    ], remainder="passthrough").set_output(transform="pandas")

    return ct
```

```python
knn = KNeighborsClassifier()

ct = buildColTransformer()

my_pipeline = Pipeline(
        [("preprocessing", ct),
        ("model", knn)]
).set_output(transform = "pandas")

search_dict = {"model__n_neighbors" : np.arange(1,100,2)}

grid_pipeline = GridSearchCV(my_pipeline, search_dict, cv=5, scoring="accuracy")
grid_results = grid_pipeline.fit(X, y)
search_dict_keys = [f"param_{item}" for item in search_dict.keys()]
search_dict_keys.append("mean_test_score")
grid_df = pd.DataFrame(grid_results.cv_results_)[search_dict_keys]
grid_df["mean_test_score"] = abs(grid_df["mean_test_score"])
grid_df = grid_df.sort_values(by=["mean_test_score"], ascending=False).head().reset_index(drop=True)
grid_df
```

```python
dt = DecisionTreeClassifier()

ct = buildColTransformer()

my_pipeline = Pipeline(
        [("preprocessing", ct),
        ("model", dt)]
).set_output(transform = "pandas")

search_dict = {
    "model__ccp_alpha" : [0, 0.001, 0.01, 0.1],
    "model__max_depth" : [None, 50, 60, 65, 70],
    "model__min_samples_leaf" : [25, 35, 45]
}
grid_pipeline = GridSearchCV(my_pipeline, search_dict, cv=5, scoring="accuracy")
grid_results = grid_pipeline.fit(X, y)
search_dict_keys = [f"param_{item}" for item in search_dict.keys()]
search_dict_keys.append("mean_test_score")
grid_df = pd.DataFrame(grid_results.cv_results_)[search_dict_keys]
grid_df["mean_test_score"] = abs(grid_df["mean_test_score"])
grid_df = grid_df.sort_values(by=["mean_test_score"], ascending=False).head().reset_index(drop=True)
grid_df
```

```python
final_dt = DecisionTreeClassifier(ccp_alpha=0.01, max_depth=None, min_samples_leaf=25)

my_pipeline = Pipeline(
        [("preprocessing", ct),
        ("model", final_dt)]
).set_output(transform = "pandas")


fitted_pipeline = my_pipeline.fit(X, y)
feature_names = ct.fit_transform(X).columns

plt.figure(figsize=(10,8))
plot_tree(fitted_pipeline["model"], feature_names=feature_names, filled=True)

plt.show()
```

```python
lda = LinearDiscriminantAnalysis()

ct = buildColTransformer()

my_pipeline = Pipeline(
        [("preprocessing", ct),
        ("model", lda)]
).set_output(transform = "pandas")

search_dict = {
    "model__solver" : ["svd", "lsqr"]
}

grid_pipeline = GridSearchCV(my_pipeline, search_dict, cv=5, scoring="accuracy")
grid_results = grid_pipeline.fit(X, y)
search_dict_keys = [f"param_{item}" for item in search_dict.keys()]
search_dict_keys.append("mean_test_score")
grid_df = pd.DataFrame(grid_results.cv_results_)[search_dict_keys]
grid_df["mean_test_score"] = abs(grid_df["mean_test_score"])
grid_df = grid_df.sort_values(by=["mean_test_score"], ascending=False).head().reset_index(drop=True)
grid_df
```
