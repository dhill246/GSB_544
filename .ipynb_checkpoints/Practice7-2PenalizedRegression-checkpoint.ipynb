```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import r2_score
```

```python
# Read the data
ames = pd.read_csv("https://www.dropbox.com/scl/fi/g0n5le5p6fr136ggetfsf/AmesHousing.csv?rlkey=jlr9xtz1o6u5rghfo29a5c02f&dl=1")

# Get rid of columns with mostly NaN values
good_cols = ames.isna().sum() < 100
ames = ames.loc[:,good_cols]

# Drop other NAs
ames = ames.dropna()
```

#### Data Cleaning
- Adjust categorical columns to actually be objects instead of numerical data types
- Adjust numerical columns to be numerical data types instead of objects or other data types
- Handle missing data (do not fill with zeroes)
  - We will drop whole columns if they have greater than 100 NAs
  - We will drop individuals rows in the rest of the dataset that have NAs

```python
ames[["Order", "PID"]]
```

```python
# If there are variables we know we never want to use, we need to explicitly
# drop them so that we can use remainder = "passthrough" in the column
# transformer.

X = ames.drop(["SalePrice", "Order", "PID"], axis = 1)
y = ames["SalePrice"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline_1 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)
```

#### Descriptions of new things:
- make_column_selector(dtype_include=object): Just find all the columns where the type is object
- make_column_selector(dtype_include=np.number): Just find all the columns where the type is numeric
- handle_unknown="ignore" in the OneHotEncoder is happening because there are categorical variables that appear in the test data but they were not seen by the training data
- We drop the Order and PID columns because they are identifier columns, which we always drop before modeling.



```python
cross_val_score(lr_pipeline_1, X, y, cv = 5, scoring = 'r2')
```

#### We used so many predictors that we overfit the training data way too much and ended up with terrible predictions

### Ridge Regression
- OLS is too flexible when we have a lot of betas
- Regularization: Change our idea of best betas. We are now going to restrict our definition of a best beta are minimizing a sum of squared residuals while also minimizing the size of the betas themselves
- The ridge penalty added to the sum of squared residuals introduces a new term operating against the other.
- We have to decide how important the residuals being small vs betas being small is. How much restriction λ should we introduce? If λ is zero, we are back to ols, if it is big, we won't care as much about residuals.



```python
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

ridge_pipeline_1 = Pipeline(
  [("preprocessing", ct),
  ("ridge_regression", Ridge(alpha=1))]
)
```

```python
cross_val_score(ridge_pipeline_1, X, y, cv = 5, scoring = 'r2')
```

### Practice Activity: Ridge Regression

Fit the model on the whole dataset and get the coefficients. Make a plot of these coefficients compared to the ones from ordinary linear regression.

```python
fitted_lr_pipeline = lr_pipeline_1.fit(X, y)
fitted_ridge_pipeline = ridge_pipeline_1.fit(X, y)
```

```python
lr_coefs = fitted_lr_pipeline.named_steps["linear_regression"].coef_
ridge_coefs = fitted_ridge_pipeline.named_steps["ridge_regression"].coef_
```

```python
coefs_df = pd.DataFrame({"LinearRegressionCoefs" : lr_coefs,
                         "RidgeRegressionCoefs" : ridge_coefs})
```

```python
from plotnine import ggplot, aes, geom_point
```

```python
(ggplot(coefs_df, aes(x="LinearRegressionCoefs", y="RidgeRegressionCoefs"))
+ geom_point()
)
```

### Practice Activity: Tuning
- We don't really know without trying many possibilities for λ what the best value is.
- λ is a hyperparameter, which means we can perform tuning on it.

Using the same pipeline as previously, perform tuning on λ

You should always try λ values on a log scale; that is, don’t use [1,2,3,4]; instead use something like [0.001, 0.01, 0.1, 1, 10]

```python
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

ridge_pipeline_grid = Pipeline(
  [("preprocessing", ct),
  ("ridge_regression", Ridge())]
)
```

```python
alphas = {"ridge_regression__alpha" : [0.1, 1, 10, 100, 1000]}
gscv = GridSearchCV(ridge_pipeline_grid, alphas, cv = 5, scoring="r2")
```

```python
gscv_fitted = gscv.fit(X, y)
```

```python
ridge_df = pd.DataFrame(data = {"alphas": [0.1, 1, 10, 100, 1000], "scores": gscv_fitted.cv_results_['mean_test_score']})
```

```python
ridge_df.sort_values(by=["scores"], ascending=False)
```

```python
# Alpha 10 Ridge:
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

ridge_pipeline_10 = Pipeline(
  [("preprocessing", ct),
  ("ridge_regression", Ridge(10))]
)
```

```python
fitted_ridge_pipeline_10 = ridge_pipeline_10.fit(X, y)
```

```python
coefs_df["Ridge_10"] = fitted_ridge_pipeline_10["ridge_regression"].coef_
```

#### Lasso Regression
- Now we want to use an absolute value instead of a squared value for our beta penalizer.
- Ridge Regression will result in coefficients that are generally *smaller* than OLS, Lasso will cause some coefficients to be *equal* to zero, which is essentially **automatic feature selection**.

### Practice Activity: Lasso

Create a LASSO pipeline, and tune λ.

Fit your best model on the full Ames data, and compare the coefficients to Ridge and OLS

```python
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lasso_pipeline_grid = Pipeline(
  [("preprocessing", ct),
  ("lasso_regression", Lasso())]
)
```

```python
alphas = {"lasso_regression__alpha" : [1, 10, 100, 1000]}
gscv = GridSearchCV(lasso_pipeline_grid, alphas, cv = 5, scoring="r2")
```

```python
lasso_df = pd.DataFrame(data = {"alphas": [1, 10, 100, 1000], "scores": gscv_fitted.cv_results_['mean_test_score']})
```

```python
lasso_df.sort_values(by=["scores"], ascending=False)
```

```python
# Alpha 100 Ridge
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lasso_pipeline_100 = Pipeline(
  [("preprocessing", ct),
  ("lasso_regression", Lasso(alpha=100))]
)
```

```python
fitted_lasso_pipeline_100 = lasso_pipeline_100.fit(X, y)
```

#### Practice Activity: Elastic Net
- We can add both ridge and lasso penalties to our model.
- We can create an elastic net pipeline, and tune λ and α
- α is actually called l1_ratio in scikit learn, and λ is called alpha.

```python
# ElasticNet

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

elastic_net = Pipeline(
  [("preprocessing", ct),
  ("elastic_net", ElasticNet())]
)
```

```python
alphas_l1_ratios = {"elastic_net__alpha" : [0.01, 0.1, 1, 10],
                    "elastic_net__l1_ratio" : [0.25, 0.5, 0.75]}
gscv = GridSearchCV(elastic_net, alphas_l1_ratios, cv = 5, scoring="r2")
```

```python
gscv_fitted = gscv.fit(X, y)
```

```python
final_results = pd.DataFrame(gscv_fitted.cv_results_)[["param_elastic_net__alpha", "param_elastic_net__l1_ratio", "mean_test_score"]]
final_results.sort_values(by=["mean_test_score"], ascending=False)
```

```python
# Final elastic net

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

elastic_net = Pipeline(
  [("preprocessing", ct),
  ("elastic_net", ElasticNet(alpha=0.01, l1_ratio=0.5))]
)

fitted_elastic_net = elastic_net.fit(X, y)
```

```python
coefs["Elastic_Net"] = fitted_elastic_net["elastic_net"].coef_
```

```python
coefs
```

```python
(ggplot(coefs, aes(x="index"))
+ geom_point(aes(y="Elastic_Net"), color="red")
+ geom_point(aes(y="Lasso_100")))
```

```python
(ggplot(coefs, aes(x="index"))
+ geom_point(aes(y="Elastic_Net"), color="red")
+ geom_point(aes(y="Ridge_10")))
```

##### Coefficient Interpretation
- The optimized elastic net with alpha=0.01 and l1_ratio=0.5 has coefficients spaced out more than the lasso model with alpha=100.
- The optimized elastic net with alpha=0.01 and l1_ratio=0.5 has coefficients which are extremely close to all the ridge regression model. This is interesting because the l1_ratio is 0.5, so I would have expected something more in between lasso and ridge instead of so close to ridge."
